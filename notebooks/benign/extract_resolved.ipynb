```python
import pandas as pd
import os

chunks_folder = '../../datasets/alexa-top-1m/chunks'
chunks_2nd_attempt_folder = '../../datasets/alexa-top-1m/chunks/2nd_attempt'
chunks_with_vpn_folder = '../../datasets/alexa-top-1m/chunks/2nd_attempt/with_vpn'
chunks = []

for file_path in os.listdir(chunks_folder):
  if file_path.endswith('.csv'):
    chunks.append(pd.read_csv(f'{chunks_folder}/{file_path}', index_col=0))
    
for file_path in os.listdir(chunks_2nd_attempt_folder):
  if file_path.endswith('.csv'):
    chunks.append(pd.read_csv(f'{chunks_2nd_attempt_folder}/{file_path}', index_col=0))
    
for file_path in os.listdir(chunks_with_vpn_folder):
  if file_path.endswith('.csv'):
    chunks.append(pd.read_csv(f'{chunks_with_vpn_folder}/{file_path}', index_col=0))
```

```python
for chunk in chunks:
  print(chunk.shape)
```

```python
chunks[0]
```

```python
df = pd.concat(chunks, ignore_index=True).sort_values(by='resolved_url').drop_duplicates(subset='url', ignore_index=True)

df.info()
print(df.describe())
df
```

```python
na_mask = df['resolved_url'].isna()

na_mask.sum()
```

```python
resolved_df = df.loc[~na_mask]

resolved_df
```

```python
resolved_df.info()
resolved_df.describe()
```

```python
resolved_df[['resolved_url']].drop_duplicates(ignore_index=True).rename(columns={'resolved_url': 'url'}).to_csv(f'../../datasets/alexa-top-1m/top-1m-resolved.csv', )
```
